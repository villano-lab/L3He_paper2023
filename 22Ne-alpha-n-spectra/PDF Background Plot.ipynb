```python
def normal_distribution(E,E1,o): 
    return N*(np.exp(-((E - E1)**2) / (2 * (o**2))))
```

```python
N= 1
E= 10
E1= 10
o= 1 

#E_range=np.linspace(0,100,101) 
E_range = np.linspace(E1 - 5 * o, E1 + 5 * o, 1000)
#print(E_range)
```

```python
pdf_values = normal_distribution(E_range,E1, o) 

plt.plot(E_range, pdf_values)
plt.xlabel('Energy (E)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Function')
plt.legend()
plt.grid(True)
plt.show()
```
