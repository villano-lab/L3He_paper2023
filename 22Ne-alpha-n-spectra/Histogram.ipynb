```python
import numpy as np
from sympy import symbols
import math
import cmath 

import matplotlib as mpl
from matplotlib import pyplot as plt 

from scipy.stats import norm
import seaborn as sb
```

```python
#Histogram 
num_samples=1000 

U= np.random.uniform(0, 1, num_samples) 

mean_energy = 0.5
std_deviation = 0.1
N=np.random.normal(mean_energy, std_deviation, num_samples) 


fig,axes = plt.subplots(1,1,figsize=(10.0,10.0),sharex=True)
plt.hist(U, bins=20, color='blue', alpha=0.7)

plt.hist(N, bins=20, color='green', alpha=0.7) 

ymin = 0
ymax = 150

plt.title('Random Normal and Uniform Distribution')
plt.xlabel('Energy (MeV)')
plt.ylabel('Frequency') 

plt.show()
```
