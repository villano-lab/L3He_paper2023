```python
# Combined graphs 

f = open('data/JENDL-files/jendl_3He.endf')
lines = f.readlines()
sec = ENDF6.find_section(lines, MF=3, MT=1)  # TOTAL
x, y = ENDF6.read_table(sec) 

sec2 = ENDF6.find_section(lines, MF=3, MT=2)  # Elastic 
x2, y2 = ENDF6.read_table(sec2)

sec3 = ENDF6.find_section(lines, MF=3, MT=102)  # CAPTURE
x3, y3 = ENDF6.read_table(sec3) 

sec4 = ENDF6.find_section(lines, MF=3, MT=103)  #(N,P) 
x4, y4 = ENDF6.read_table(sec4)


#plot quick
fig = plt.figure()
ax1 = fig.add_subplot(111)
plot(x, y,color='black',label='Total') 
plot(x2, y2,color='r',label='Elastic') 
plot(x3, y3,color='orange', linestyle='dashed',label='Capture') 
plot(x4, y4,color='deepskyblue',linestyle='dashed',label='(n,p)')
ax1.set_yscale('log')
ax1.set_xscale('log') 
title('n + $^3\mathrm{He}$')
xlabel('Neutron energy [eV]')
ylabel('Cross-section [barn]')
#plt.savefig('3Hexn.png') 

legend()
show()
```
