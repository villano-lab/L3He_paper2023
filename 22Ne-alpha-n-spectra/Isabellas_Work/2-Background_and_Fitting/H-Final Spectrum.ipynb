```python
import numpy as np
from sympy import symbols
import math
import cmath 

import matplotlib as mpl
from matplotlib import pyplot as plt 

from scipy.stats import norm
import seaborn as sb 

import lmfit as lf
import scipy.optimize as opt 

from scipy.optimize import curve_fit  

from scipy import special
```

```python
# Functions for the Helium Process
def normal_distribution(E,N,E1,o): 
    return N*(np.exp(-((E - E1)**2) / (2 * (o**2)))) 

def error_function(x): 
    return special.erfc(((x-5)/2)*3)

#Values for the helium process
N= 1
E= 10
E1= 10
o= 1 

N2= 1
EE2= 4
E2= 4
o2= 1 

#Energy range
#E_range=np.linspace(0,100,101) 
E_range = np.linspace(0, 20, 1000)
#print(E_range)
```

```python
plt.plot(E_range, normal_distribution(E_range,N,E1,o)+normal_distribution(E_range,N2,EE2,o2)+error_function(E_range))
plt.xlabel('Energy (E)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Potassium Background 
math.cos 
#print(math.cos)
np.pi 

num_points = 1000 
a_values = np.linspace(0, np.pi, num_points) 
cos_values = np.cos(a_values) 

#Compton Function λ′ − λ = (h/mc)(1 − cos θ)
def compton_formula(h,m,c):
    return (h/m*c)*(1-(cos_values))

h = 6.62607015*10**-34 #(J⋅s) 
m = 0.511 #MeV/c² 
c = 3*10**8 

#print(compton_formula(h,m,c))
```

```python
plt.plot(cos_values,compton_formula(h,m,c))
```

```python
plt.plot(E_range, normal_distribution(E_range,N,E1,o)+normal_distribution(E_range,N2,EE2,o2)+error_function(E_range)+compton_formula(h,m,c))
plt.xlabel('Energy (E)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Function')
plt.legend()
plt.grid(True)
plt.show()
```
