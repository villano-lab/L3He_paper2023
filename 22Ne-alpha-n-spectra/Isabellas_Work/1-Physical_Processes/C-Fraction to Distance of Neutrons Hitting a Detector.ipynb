```python
#Distance r 
r=np.linspace(1,10,10) 
print (r)
```

```python
#all SA with r  
SA= 4*math.pi*r**2
print (SA)
```

```python
#Fractrion of Neutrons hitting for all
F=(DA/SA)*100 
print(F)
```

```python
#Plot 
fig,axes = plt.subplots(1,1,figsize=(15.0,8.0),sharex=True)
ax1 = axes 
F=np.asarray (F) 
r=np.asarray (r)
ax1.plot(F,label='F,r  calculation') 

ymin = 0
ymax = 33


plt.grid()
#ax1.set_yscale('linear')
#ax1.set_yscale('log')
ax1.set_xlim(0, 10) 
ax1.set_ylim(ymin,ymax) 
ax1.set_title(r'Fraction of Neutrons Hitting at Range of Distances', fontsize=20) 
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax1.set_xlabel(r'Distance(cm)', fontsize=20)
ax1.set_ylabel('Percentage of Neutrons Hitting', fontsize=20)
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')
ax1.legend(loc=1,prop={'size':22})
#ax1.legend(bbox_to_anchor=(1.04,1),borderaxespad=0,prop={'size':22})

for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
#plt.savefig('paper_figures/hist.eps')
#plt.savefig('paper_figures/hist.pdf') 
#plt.savefig('1CPlot1.png')
plt.show()
```

```python
#Counts 
# Detection efficiency values (0% to 100%)
detection_efficiency = np.arange(0, 101, 10)

# Values
pb = 10
uA = 100
days = 100
factor_30 = 0.0269 * .30 * pb * uA * days
factor_3 = 0.0269 * .03 * pb * uA * days

# Calculate expected detected counts for 30% and 3% geometric coverage
counts_30 = factor_30 * detection_efficiency/100
counts_3 = factor_3 * detection_efficiency/100 
print(counts_30)
print(counts_3)
```

```python
#Factor depending on distance 
# Values
pb = 10
uA = 100
days = 100
factor= 0.0269 * (F/100) * pb * uA * days 
print(factor)
```

```python
#Counts with efficiency at .154 
counts=factor * .154 
print(counts)
```
